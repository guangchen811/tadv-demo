```python
from tadv.loader import FileLoader
import pandas as pd
df = FileLoader.load_csv("./imdb_top_1000.csv")
```

```python
year_str = df['Released_Year'].astype(str).str.strip()
year_int = pd.to_numeric(year_str, errors='coerce')
df = df[year_int.notna()]
```

```python
df.to_csv("./imdb_top_1000_clean.csv", index=False, encoding="UTF-8")
```
