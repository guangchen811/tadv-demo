```python
import pandas as pd

df = pd.read_csv("./data.csv", encoding="UTF-8", sep=";")
```

```python
c2_woe = 'Curricular units 2nd sem (without evaluations)'
c2_enr = 'Curricular units 2nd sem (enrolled)'
c2_app = 'Curricular units 2nd sem (approved)'
c2_cred = 'Curricular units 2nd sem (credited)'
c_app_order = 'Application order'
df = df.loc[df[c2_woe] <= df[c2_enr]].reset_index(drop=True)
df = df.loc[df[c_app_order] >= 1].reset_index(drop=True)

mask = df[c_app_order] >= 1
violating_rows = df.loc[~mask]

for _, row in violating_rows.iterrows():
    print(",".join(str(v) for v in row))
    break
```

```python
df.to_csv("./students_data_clean.csv", index=False, encoding="UTF-8")
```
