```python
from tadv.loader import FileLoader
import pandas as pd
df = FileLoader.load_csv("./matches.csv")
```

```python
df['season'] = pd.to_numeric(df['season'], errors='coerce').astype('Int64')
df = df[df['season'].notna()]
df['season'].notna().sum() / df['season'].notna().count()
```

```python
df.to_csv("./matches_clean.csv", index=False, encoding="UTF-8")
```
